**Objective:** Install Docker, write a simple Python application, and create a `Dockerfile` to containerize and run it.

### Step 1: Install Docker
First, we update our package manager and install the Docker engine and command-line interface (CLI) tools.

```python
!apt-get update -qq
!apt-get install -y docker.io > /dev/null
!docker --version
```

### Step 2: Create the Application
Before we can containerize an application, we need an application to run! We will create a very simple Python script that acts as our app. It will live in a new directory called `my_app`.

```python
# Create a directory for our project
!mkdir -p my_app
```

```python
%%writefile my_app/app.py
def main():
    print("--------------------------------------------------")
    print("🚀 Hello from inside the Docker Container!")
    print("✅ Application successfully containerized and running.")
    print("--------------------------------------------------")

if __name__ == "__main__":
    main()
```

### Step 3: Create the Dockerfile
The `Dockerfile` is a set of instructions that tells Docker how to build the image.
Our instructions will:
1. Pull a lightweight official Python image (`python:3.9-slim`).
2. Set a working directory inside the container (`/app`).
3. Copy our `app.py` file into the container.
4. Tell the container to run the Python script when it starts.

```python
%%writefile my_app/Dockerfile
# Use an official Python runtime as a parent image
FROM python:3.9-slim

# Set the working directory inside the container
WORKDIR /app

# Copy the current directory contents into the container at /app
COPY app.py .

# Command to run the application
CMD ["python", "app.py"]
```

### Step 4: Build and Run the Docker Image
With our application and Dockerfile ready, we use the `docker build` command to package our app into an image named `my-python-app`. Finally, we use `docker run` to execute it.

*(Note: In a standard local or cloud VM environment, the following commands execute the container. In restricted notebook environments, the daemon execution is blocked for security, but the syntax remains exactly the same).*

```python
# Navigate to the app directory and build the Docker image
!cd my_app && docker build -t my-python-app .

# Run the containerized application
!docker run my-python-app
```
